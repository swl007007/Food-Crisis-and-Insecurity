```python
import numpy as np
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import new_merge.csv
df_new = pd.read_csv('new_merged_dataset.csv')
```

```python
df_new['date'] = pd.to_datetime(df_new['date'])
```

```python
df_new['country_en'].unique()
```

```python
# import pycountry
import pycountry
```

```python
# from two letter country code to three letter country code
def country_code_convert(country):
    try:
        return pycountry.countries.get(alpha_2=country).alpha_3
    except:
        return np.nan

# apply the function to the country_code column
df_new['country_code_3'] = df_new['country_code'].apply(country_code_convert)
```

```python
# set Namibia country_code_3 to NAM
df_new.loc[df_new['country_en'] == 'Namibia', 'country_code_3'] = 'NAM'
```

```python
# import CPI.csv GDP.csv
df_cpi = pd.read_csv(r'C:\Users\WeilunShi\OneDrive - CGIAR\Desktop\Food Crisis and Insecurity\0.external_data\WorldBank\CPI.csv')
df_gdp = pd.read_csv(r'C:\Users\WeilunShi\OneDrive - CGIAR\Desktop\Food Crisis and Insecurity\0.external_data\WorldBank\GDP.csv')
```

```python
# remove indicator_name column
df_cpi = df_cpi.drop(columns=['Indicator Name'])
# reshape df_cpi
df_cpi = df_cpi.melt(id_vars=['Country Name', 'Country Code'], var_name='year', value_name='CPI')
# remove Country Name column, rename Country Code to country_code_3
df_cpi = df_cpi.drop(columns=['Country Name'])
# rename Country Code to country_code_3
df_cpi = df_cpi.rename(columns={'Country Code': 'country_code_3'})
# extract year from df_new
df_new['year'] = df_new['date'].dt.year
# convert df_cpi['year'] to int
df_cpi['year'] = df_cpi['year'].astype(int)
# merge df_new and df_cpi basd on country_code_3 and year, add indicator
df_new_cpi = pd.merge(df_new, df_cpi, on=['country_code_3', 'year'], how='left',indicator=True)
df_new_cpi['_merge'].value_counts()
# drop _merge column
df_new_cpi = df_new_cpi.drop(columns=['_merge'])
```

```python
# remove indicator_name column
df_gdp = df_gdp.drop(columns=['Indicator Name'])
# reshapgdp
df_gdp = df_gdp.melt(id_vars=['Country Name', 'Country Code'], var_name='year', value_name='GDP')
# remove Country Name column, rename Country Code to country_code_gdp 
df_gdp = df_gdp.drop(columns=['Country Name'])
# rename Country Code to country_code
df_gdp = df_gdp.rename(columns={'Country Code': 'country_code_3'})
# convergdp['year'] to 
df_gdp['year'] = df_gdp['year'].astype(int)
# merge df_new angdp basd on country_code_3 and year, add indicator
df_new_gdp = pd.merge(df_new_cpi,df_gdp, on=['country_code_3', 'year'], how='left',indicator=True)


df_new_gdp['_merge'].value_counts()


# drop _merge column
df_new_gdp = df_new_gdp.drop(columns=['_merge'])
```

```python
# import infra_index
df_infra = pd.read_csv(r'C:\Users\WeilunShi\OneDrive - CGIAR\Desktop\Food Crisis and Insecurity\0.external_data\WorldBank\infra_index.csv')
# drop Indicator Code
df_infra = df_infra.drop(columns=['Indicator Code'])
df_infra = df_infra.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name'], var_name='year', value_name='Value')
```

```python
# filter year from 2017 to 2022
df_infra = df_infra[df_infra['year'].isin(['2017', '2018', '2019', '2020', '2021', '2022'])]
```

```python

# see missing percentage divided by specific Indicator Name
df_infra.groupby('Indicator Name').apply(lambda x: x.isnull().sum()/len(x))
```

```python
# extract Liner shipping connectivity index (maximum value in 2004 = 100) from df_infra
df_infra_index = df_infra[df_infra['Indicator Name'] == 'Liner shipping connectivity index (maximum value in 2004 = 100)']
```

```python
# drop Indicator Name
df_infra_index = df_infra_index.drop(columns=['Indicator Name'])
```

```python
# remove Country Name column, rename Country Code to country_code_gdp 
df_infra_index = df_infra_index.drop(columns=['Country Name'])
# rename Country Code to country_code
df_infra_index = df_infra_index.rename(columns={'Country Code': 'country_code_3'})
# convergdp['year'] to 
df_infra_index['year'] = df_infra_index['year'].astype(int)
# merge df_new angdp basd on country_code_3 and year, add indicator
df_infra_index_new = pd.merge(df_new_gdp,df_infra_index, on=['country_code_3', 'year'], how='left',indicator=True)

```

```python

df_infra_index_new['_merge'].value_counts()
```

```python
# drop _merge column
df_infra_index_new = df_infra_index_new.drop(columns=['_merge'])
```

```python
# rename value to infra_index
df_infra_index_new = df_infra_index_new.rename(columns={'Value': 'infra_index'})
```
