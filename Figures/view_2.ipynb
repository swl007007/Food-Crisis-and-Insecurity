By year breakdown, to check battle data is integrated into the IPC dataset.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Read the GeoJSON file
geojson_file = "output.geojson"
gdf = gpd.read_file(geojson_file)
gdf.year = gdf.year.astype(int)

# Reproject the data to Web Mercator (EPSG:4326) coordinate reference system
gdf = gdf.to_crs(epsg=4326)
years = list(range(2017,2023))

# Choose a specific property to visualize
selected_property = 'battles'  # Replace with the property name you want to visualize

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 15))

# Iterate over the years and create a subplot for each year
for i, year in enumerate(years):
    ax = axes[i // 2, i % 2]

    # Filter the data for the current year
    filtered_gdf = gdf[gdf['year'] == year]  # Replace 'year' with the correct column name for the year

    # Create a choropleth map based on the selected property
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)
    filtered_gdf.plot(column=selected_property, cmap='YlOrRd', linewidth=0.1, edgecolor='black', legend=True, ax=ax, cax=cax)

    # Add a basemap
    ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)

    # Set the title of the map
    ax.set_title(f'Choropleth Map: {selected_property} ({year})', fontsize=16)

    # Remove axis ticks and labels
    ax.set_xticks([])
    ax.set_yticks([])

# Adjust the layout to avoid overlapping plots
fig.tight_layout()

# Show the map
plt.show()
```
